```python
# Identificar columnas booleanas automáticamente
bool_cols = df.select_dtypes(include=['bool']).columns

# Convertir a 0 y 1
df[bool_cols] = df[bool_cols].astype(int)

# Ver resultado
print(df[bool_cols].head())
```

```python
df['Duration_sec'] = df['Length'].str.split(':').apply(lambda x: int(x[0])*60 + int(x[1]))
df.drop(columns=['Length'], inplace=True)  # Opcional: eliminar la columna original
```

```python
df['Loudness (db)'] = pd.to_numeric(df['Loudness (db)'], errors='coerce')
```
